### MP NEuRON Class

```python
class MPNeuron:
  
  def __init__(self):          # Declaring b 
    self.b = None
    
  def model(self, x):          # Finding if the predicted vallue of b, satifies the condition 
    return(sum(x)>= self.b)
  
  def predict(self,x):         # Predicting the number of conditions that are statisfied for the value of 'b'. 
    y = []
    for x in x:
      result = self.model(x)
      y.append(result)
    return np.array(y)
  
  def fit(self, x,y):
    accuracy = {}
    for b in range(x.shape[1]+1):
      self.b = b  
      y_pred = self.predict(x)
      accuracy[b] = accuracy_score(y_pred,y)
    best_b = max(accuracy, key = accuracy.get)
    self.b = best_b
    print('Optimal value of b:', best_b)
    print('highest accuracy is', accuracy[best_b])
```

## Perceptron Class

$y=1, \mbox{if}  \sum_i w_i x_i>=b$|

$ y = 0, \mbox{otherwise}$

```python
X_train = x_train.values
X_test = x_test.values
```

```python
# The model assumes if the sum of the numerical(float) values of the input along with their weights satifies the condition of being greater than b, then it is else 0
# This assumption is then compared with the actual result to determine the nature of prediction of the predicted value
class Perceptron:
  
  def __init__ (self):
    self.w = None
    self.b = None
                                                              ## Function to determine if the weight(w)*input(x) and see if it satifies the condition of being greater than the threshold b  
  def model(self,x):
    return 1 if (np.dot(self.w,x)>= self.b) else 0
                                                              ## This function is called after value of b and the weights are determined to create an array of the predicted output  
  def predict(self,X):
    y=[]
    for x in X: 
      result = self.model(x)
      y.append(result)
    return np.array(y)
                                                               ## The fit function is run to predict the right
  def fit(self, x, y, epochs =1, lr= 1):
    self.w = np.ones(x.shape[1])                                # Predifining the values of the indivdual weights as "1"
    self.b = 0                                                  # Predifining the value of b as "0"
    accuracy = {}                                               # Creating the accuracy set to store the prediction result and accuracy 
    max_accuracy = 0                                            # Predifining max_accuracy as 0 
    for i in range(epochs):                                     # Running i through the epochs deined by the devoloper, to increase the accuracy(also increases time for training)
      for X,Y in zip(x,y):                                      # Extracting the individual array(row) from the train data
        y_pred = self.model(X)                                  # Finding the value of y_pred for the row selected 
                                                               ## If the predicted values does not  
        if Y==1 and y_pred == 0:                                # If y_pred is 0 and Y is 1, it means our value of the weights.inputs is lesser than the value b
          self.w = self.w+X                                     # Here we increase the 
          self.b = self.b+1
        elif Y==0 and y_pred == 1:
          self.w =self.w - lr*X
          self.b = self.b - lr*1
      accuracy[i] = accuracy_score(self.predict(x),y)
      if (accuracy[i]>max_accuracy):
        max_accuracy = accuracy[i]
        chkptw = self.w
        chkptb = self.b
        
    self.w = chkptw 
    self.b = chkptb
     
    print(max_accuracy)
      
    plt.plot(accuracy.values())
    plt.show()
```

```python
def calculate_loss(X,Y,w_est,b_est):
  loss = 0 
  for x,y in zip(X,Y):
    loss += (y-sigmoid(x,w_est,b_est))**2
  return loss 

def sigmoid_loss(x,y,w,b):
  P_loss = np.zeros(w.shape)
  for i in range(w.shape[0]):
    for j in range(w.shape[1]):
      P_loss[i,j] = calculate_loss(x,y,w[i,j],b[i,j])
  return P_loss
```

## Sigmoid Neuron Class

```python
# code to find the sigmoid function of matrices
def sigmoid_arr(x, w, b):
  return 1.0/(1.0+ np.exp(-(np.dot(x,w)+b)))
```

### Class Code 

```python
class SigmoidNeuron:
  
  def __init__(self):
    self.w = None 
    self.b = None 
    
  def perceptron(self,x):
    return np.dot(x,self.w.T)+self.b
  
  def sigmoid(self, x):
    return 1.0/(1.0+np.exp(-x))
  
  def grad_w(self, x, y):
    fx = self.sigmoid(self.perceptron(x))
    return (fx - y)*fx*(1-fx)*x
  
  def grad_b(self, x, y):
    fx = self.sigmoid(self.perceptron(x))
    return (fx - y)*fx*(1-fx)
  
  def fit(self, X, Y, ephocs = 1, eta=1, initialize = True):
    # innitializing w,b 
    if initialize:
      self.w = np.random.randn(1, X.shape[1])
      self.b = 0
    for i in range(ephocs):
      dw = 0 
      db = 0 
      
      for x,y in zip(X,Y):
        dw += self.grad_w(x,y)
        db += self.grad_b(x,y)
        
      self.w -= eta*dw
      self.b -= eta*db 
```

### Toy Data (*test-1*) 2D 

```python
class Sigmoid_N2D:
  def __init__(self):
    self.w = None 
    self.b = None 
    
  def line_equ(self,x):
    return (x*self.w)+self.b
  
  def sigmoid(self, x):
    return 1.0/(1.0+np.exp(-x))
  
  def grad_w(self, x, y):
    fx = self.sigmoid(self.line_equ(x))
    return (fx - y)*fx*(1-fx)*x
  
  def grad_b(self, x, y):
    fx = self.sigmoid(self.line_equ(x))
    return (fx - y)*fx*(1-fx)
  
  def fit(self, X, Y, ephocs = 1, eta=1, initialize = True):
    # innitializing w,b 
    if initialize:
      self.w = np.random.random()
      self.b = 0
    for i in range(ephocs):
      dw = 0 
      db = 0 
      
      for x,y in zip(X,Y):
        dw += self.grad_w(x,y)
        db += self.grad_b(x,y)
        
      self.w -= eta*dw
      self.b -= eta*db 
```

```python
my_cmap = matplotlib.colors.LinearSegmentedColormap.from_list("",['red','yellow','green'])
```

```python
X = np.asarray([[2.5, 2.5], [4, -1], [1, -4], [-3, 1.25], [-2, -4], [1, 5]])
Y = [1, 1, 1, 0, 0, 0]
```

```python
sn = SigmoidNeuron()
sn.fit(X, Y, 1, 0.25, True)
```

```python
def plot_sn(X, Y, sn, ax):
  X1 = np.linspace(-10, 10, 100)
  X2 = np.linspace(-10, 10, 100)
  XX1, XX2 = np.meshgrid(X1, X2)
  YY = np.zeros(XX1.shape)
  for i in range(X2.size):
    for j in range(X1.size):
      val = np.asarray([X1[j], X2[i]])
      YY[i, j] = sn.sigmoid(sn.perceptron(val))
  ax.contourf(XX1, XX2, YY, cmap=my_cmap, alpha=0.6)
  ax.scatter(X[:,0], X[:,1],c=Y, cmap=my_cmap)
  ax.plot()
```

```python
sn.fit(X, Y, 1, 0.05, True)
N = 30
plt.figure(figsize=(10, N*5))
for i in range(N):
  print(sn.w, sn.b)
  ax = plt.subplot(N, 1, i + 1)
  plot_sn(X, Y, sn, ax)
  sn.fit(X, Y, 1, 0.5, False)
```
