```python
# function that plots vectors
```

```python
def vector_plot(a): 
    import numpy as np
    import matplotlib.pyplot as plt
    
    colors = ['r','y','g','b']
    l = 0 
    for i in a: 
        plt.quiver(i[0],i[1],i[2],i[3], scale_units='xy', angles='xy', scale=1, color = colors[l%len(colors)])
        plt.xlim(-10,10)
        plt.ylim(-10,10)
        l+=1
    plt.show()
```

```python
a = [[0,0,2,6],[1,1,4,5]]
vector_plot(a)
```
