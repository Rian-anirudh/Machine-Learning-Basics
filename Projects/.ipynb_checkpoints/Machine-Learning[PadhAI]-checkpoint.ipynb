## Debuging

```python
import random
```

```python
def factorial(x):
    if (x==0):
     return 1 
    return x*factorial(x-1)
```

```python
factorial(50)
```

```python
def code_to_debug(): 
    import pdb; pdb.set_trace()
    
    for i in range(10):
        x = random.random()
        factorial(x)
```

```python
%xmode Verbose
```

```python
code_to_debug()
```
